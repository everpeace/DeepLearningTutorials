# Denoising Auto Encoder(dA) 評価(Eval) スクリプト
## dA 復元 (層構成: 28*28 --> 500)

```python
weight_file_path='dA_plots/weights'
dataset='mnist.pkl.gz'

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
%matplotlib inline
import numpy
import theano
import theano.tensor as T
import os
from theano.tensor.shared_randomstreams import RandomStreams
from utils import tile_raster_images
from logistic_sgd import load_data
from dA import dA

# Random Generators
rng = numpy.random.RandomState(123)
theano_rng = RandomStreams(rng.randint(2 ** 30))

# dA復元
x = T.matrix('x')  # the data is presented as rasterized images
da = dA(
            numpy_rng=rng,
            theano_rng=theano_rng,
            input=x,
            n_visible=28 * 28,
            n_hidden=500
        )

# 重み読み込み
import cPickle
weights = open(weight_file_path)
for p in da.params:
    p.set_value(cPickle.load(weights), borrow=True)
weights.close()

# DA 評価関数(類似入力を思い出させる)
vis = T.matrix('vis')
da_eval = theano.function([vis], 
                                              da.get_reconstructed_input(da.get_hidden_values(da.get_corrupted_input(vis, 0.3))),
                                              name='da_eval')

# データセット読み込み
datasets = load_data(dataset)
train_set_x, train_set_y = datasets[0]
valid_set_x, valid_set_y = datasets[1]
test_set_x, test_set_y = datasets[2]
```

## 復元されたdAの隠れ層の表現

```python
plt.imshow(tile_raster_images( X=da.W.get_value(borrow=True).T,
                                                       img_shape=(28, 28), tile_shape=(10, 10), tile_spacing=(1, 1)), 
                    cmap = cm.Greys_r)
```

## テスト入力イメージ表示と復元結果
### 入力イメージ(100例)

```python
input = numpy.array(test_set_x.get_value(borrow=True))
ans = numpy.array(test_set_y.eval())
plt.imshow(tile_raster_images(X=input[100:200], 
                                                      img_shape=(28, 28), tile_shape=(5,20), tile_spacing=(1, 1)),
                     cmap = cm.Greys_r)
```

### 復元イメージ

```python
remembered = da_eval(input)
plt.imshow(tile_raster_images(X=remembered[100:200], img_shape=(28, 28), tile_shape=(5,20), tile_spacing=(1, 1)),
                     cmap = cm.Greys_r)
```
