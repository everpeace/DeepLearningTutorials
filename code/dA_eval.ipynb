# Denoising Auto Encoder 評価(Eval) スクリプト

```python
weight_file_path='dA_plots/weights'
dataset='mnist.pkl.gz'

# RBM復元
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import numpy
import theano
import theano.tensor as T
import os
from theano.tensor.shared_randomstreams import RandomStreams
from utils import tile_raster_images
from logistic_sgd import load_data
from dA import dA

# Random Generators
rng = numpy.random.RandomState(123)
theano_rng = RandomStreams(rng.randint(2 ** 30))

# Denoising Autoencoder初期化
x = T.matrix('x')  # the data is presented as rasterized images
da = dA(
            numpy_rng=rng,
            theano_rng=theano_rng,
            input=x,
            n_visible=28 * 28,
            n_hidden=500
        )

# 重み読み込み
import cPickle
weights = open(weight_file_path)
for p in da.params:
    p.set_value(cPickle.load(weights), borrow=True)
weights.close()

# DA 評価関数(類似入力を思い出させる)
vis = T.matrix('vis')
da_eval = theano.function([vis], 
                                              da.get_reconstructed_input(da.get_hidden_values(da.get_corrupted_input(vis, 0.3))),
                                              name='da_eval')
```

```python
# データセット読み込み
datasets = load_data(dataset)
train_set_x, train_set_y = datasets[0]
valid_set_x, valid_set_y = datasets[1]
test_set_x, test_set_y = datasets[2]
```

```python
# テストデータ表示
input = test_set_x.get_value(borrow=True)[300:350]
print input.shape
plt.imshow(tile_raster_images(X=input, img_shape=(28, 28), tile_shape=(5,10), tile_spacing=(1, 1)))
```

```python
# RBMにイメージを入力し類似データを思い出させる
remembered = da_eval(input)
```

```python
# 思い出したイメージを表示
plt.imshow(tile_raster_images(X=remembered, img_shape=(28, 28), tile_shape=(5, 10), tile_spacing=(1, 1)))
```
