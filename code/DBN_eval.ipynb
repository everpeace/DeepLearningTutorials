# Deep Belief Networkの評価(Eval)スクリプト

## DBN復元 (層サイズ: (28*28) --> 1024 --> 1024 --> 1024 --> 10)

```python
weight_file_path='DBN_plots/weights'
dataset='mnist.pkl.gz'

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import matplotlib.cm as cm
%matplotlib inline
import numpy
import theano
import theano.tensor as T
import math
import sys
from theano.tensor.shared_randomstreams import RandomStreams
from utils import tile_raster_images
from logistic_sgd import load_data
from DBN import DBN
import cPickle

# Random Generators
rng = numpy.random.RandomState(123)
theano_rng = RandomStreams(rng.randint(2 ** 30))

# DBN 初期化
dbn = DBN(numpy_rng=rng, theano_rng=theano_rng,
                    n_ins=28 * 28,  hidden_layers_sizes=[1024, 1024, 1024], n_outs=10)

# 重み読み込み
weights = open(weight_file_path)
for p in dbn.params:
    p.set_value(cPickle.load(weights), borrow=True)
weights.close()

# DBN 評価関数
x = T.matrix('x')
dbn_eval = theano.function([ x ], 
                                                  dbn.logLayer.y_pred,
                                                  name='dbn_eval',
                                                  givens = {
                                                      dbn.x: x
                                                  })
# データセット読み込み
datasets = load_data(dataset)
train_set_x, train_set_y = datasets[0]
valid_set_x, valid_set_y = datasets[1]
test_set_x, test_set_y = datasets[2]
```

## テスト入力データ表示と予測結果(最初の100例/全10000例)

```python
input = numpy.array(test_set_x.get_value(borrow=True))
ans = numpy.array(test_set_y.eval())
plt.imshow(tile_raster_images(X=input[0:100], img_shape=(28, 28), tile_shape=(5,20), tile_spacing=(1, 1)), cmap=cm.Greys_r)
preds = dbn_eval(input)
print"予測結果\n"
sys.stdout.write('       ')
for i in xrange(preds[0:100].shape[0]):
    sys.stdout.write('%d     ' % preds[i])
    if (i % 20) ==19 : sys.stdout.write('\n       ')
print "\nテストイメージ" 
```

## 予測エラー(すべて)

```python
wrong_idx = numpy.where(preds!=ans)[0]
plt.imshow(tile_raster_images(X=input[wrong_idx], img_shape=(28, 28), tile_shape=(wrong_idx.shape[0]/20 + 1,20), tile_spacing=(1, 1)), 
                    cmap=cm.Greys_r)
print '予測エラー(予/実)\n'
a = ans[wrong_idx]
wrong_p = preds[wrong_idx]
sys.stdout.write('       ')
for i in xrange(wrong_idx.shape[0]):
    import sys
    sys.stdout.write('%d/%d   ' % (wrong_p[i],a[i]) )
    if (i % 20) ==19 : sys.stdout.write('\n       ')
print "\n\n予測エラーのテストイメージ"
```
