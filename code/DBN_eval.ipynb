# DBNの評価(Eval)スクリプト

## パラメータ

```python
weight_file_path='DBN_plots/weights'
dataset='mnist.pkl.gz'
```

## DBN復元

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline
import numpy
import theano
import theano.tensor as T
import os
from theano.tensor.shared_randomstreams import RandomStreams
from utils import tile_raster_images
from logistic_sgd import load_data
from DBN import DBN

# Random Generators
rng = numpy.random.RandomState(123)
theano_rng = RandomStreams(rng.randint(2 ** 30))

# DBN 初期化
dbn = DBN(numpy_rng=rng,
                    theano_rng=theano_rng,
                    n_ins=28 * 28,
                    hidden_layers_sizes=[1024, 1024, 1024],
                    n_outs=10)

# 重み読み込み
import cPickle
weights = open(weight_file_path)
for p in dbn.params:
    p.set_value(cPickle.load(weights), borrow=True)
weights.close()

# DBN 評価関数
x = T.matrix('x')
dbn_eval = theano.function([ x ], 
                                                  dbn.logLayer.y_pred,
                                                  name='dbn_eval',
                                                  givens = {
                                                      dbn.x: x
                                                  })
```
